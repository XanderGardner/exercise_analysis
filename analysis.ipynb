```python
import matplotlib.pyplot as plt

from cycle_object import cycle_object
from cycle_database import cycle_database
db = cycle_database()
```

```python
cycle_names = db.cycles_available()
cycle_names
```

```python
cycle1 = db.load(cycle_names[0])
exercise_names = cycle1.exercise_names()
num_days = len(cycle1)
```

```python
cycle1.get_volumes(exercise_names[0])
```

```python
volumes_pair = cycle1.get_set_volumes(exercise_names[0])
plt.scatter(volumes_pair[0], volumes_pair[1])
plt.title(exercise_names[0])
plt.xlabel("day")
plt.ylabel("exercise_volume")
```

```python
a = cycle1.get_zscore_set_volumes()

plt.scatter(*a, alpha=0.5)
plt.title(exercise_names[0])
plt.xlabel("day")
plt.ylabel("exercise_volume")
```

```python
a = cycle1.get_zscore_volumes()

plt.scatter(*a, alpha=0.5)
plt.title(exercise_names[0])
plt.xlabel("day")
plt.ylabel("exercise_volume")
```

```python
a = cycle1.get_zscore_one_rep_maxes()

plt.scatter(*a, alpha=0.5)
plt.title(exercise_names[0])
plt.xlabel("day")
plt.ylabel("exercise_volume")
```

```python
cycle1.exercise_names()
```

```python
for exercise in cycle1.exercise_names():
  plt.xlabel("time")
  plt.ylabel("z-scored predicted one-rep-maxes")
  plt.title("title")
  plt.plot(*cycle1.get_zscore_one_rep_maxes(exercise), label=exercise)
  plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.plot()
```

```python
for cycle_name in db.cycles_available():
  plt.xlabel("time")
  plt.ylabel("intensity")
  plt.title("intensity vs time for different exercise cycles")
  plt.plot(*db.load(cycle_name).get_avg_zscore_one_rep_maxes(), label=cycle_name)
  plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.plot()
```
